Face Detection

```python
face_cascade = cv2.CascadeClassifier(r"C:\Users\USER\Desktop\ML\image pro\Object Detection Files\haarcascade_frontalface_default.xml")
```

```python
def detect_face(img):
    
    face_img = img.copy()
   
    face_rects = face_cascade.detectMultiScale(face_img, scaleFactor = 1.2, minNeighbors = 2) 
    
    for (x,y,w,h) in face_rects: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,255,255), 2) 
        
    return face_img
```

```python
result = detect_face(vk)
```

```python
plt.imshow(result,cmap = 'gray')
```

```python
result2 = detect_face(cr)
plt.imshow(result2, cmap ='gray')
```

```python
result3 = detect_face(team)
plt.imshow(result3, cmap = 'gray')
```

Eye Detection

```python
eye_cascade = cv2.CascadeClassifier(r"C:\Users\USER\Desktop\ML\image pro\Object Detection Files\haarcascade_eye.xml")
```

```python
def detect_eyes(img):
    
    face_img = img.copy()
  
    eyes = eye_cascade.detectMultiScale(face_img, scaleFactor = 1.2, minNeighbors = 2) 
    
    
    for (x,y,w,h) in eyes: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,255,255), 3) 
        
    return face_img
```

```python
result4 = detect_eyes(vk)
plt.imshow(result4, cmap = 'gray')
```

```python
result5 = detect_eyes(cr)
plt.imshow(result5, cmap = 'gray')
```

```python
result6 = detect_eyes(team)
plt.imshow(result6, cmap = 'gray')
```
